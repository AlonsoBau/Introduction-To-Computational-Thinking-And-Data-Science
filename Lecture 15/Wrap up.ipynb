* Massachussets has area = 10,000 sq miles
* Cancer cases per year = 36,000
* Partitioned State into 1000 regions
* 10 sq miles each
* E[cases per region] = 36
* Region 111 had 143 cancer cases in 3 years (very high)

```python
random.seed(0)
numCasesPerYear = 36000
numYears = 3
stateSize = 10000
communitySize = 10
numCommunities = stateSize//communitySize
numCommunities
```

```python
numTrials = 100
numGreater = 0
for t in range(numTrials):
    locs = [0]*numCommunities # locs has one zero per community
    for i in range(numYears*numCasesPerYear): # Total number of Cases in 3 years
        locs[random.choice(range(numCommunities))] += 1
        #randomly assign one cancer case to a community
    if locs[111] >= 143: 
        numGreater += 1 # If region 111 has 143 cases or more, add 1 to numGreater
    # Repeat the process 100 times
prob = round(numGreater/numTrials, 4) # Probability of region 111 having at least 143 cases
print('Est. probability of region 111 having\
 at least 143 cases =', prob)

# But this doesn't let you evaluate if this region has something wrong with it
# This is cherry picking man
```

What is the probability of having at least 1 region that has 143?

```python
numTrials = 100
anyRegion = 0
for trial in range(numTrials):
   locs = [0]*numCommunities
   for i in range(numYears*numCasesPerYear):
       locs[random.choice(range(numCommunities))] += 1
   if max(locs) >= 143:
       anyRegion += 1 # IF a region has at least 143 cases, add 1
print(anyRegion)
aProb = round(anyRegion/numTrials, 4)
print('Est. probability of some region  having\
at least 143 cases =', aProb)
```

# PDE

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 2  # Assume L=2 for simplicity
x = 2  # Given location
t = 1  # Given time
N_max = 1000  # Maximum number of terms to sum

# Compute Fourier coefficients (assuming c_n = 1/n**2 for demonstration purposes)
def c_n(n):
    return 1 / n**3  # Example choice

# Compute the contribution of each term
n_values = np.arange(1, N_max + 1)
contributions = c_n(n_values) * np.sin(n_values * np.pi * x / L) * np.cos(n_values * np.pi * t / L)

# Compute the cumulative sum
cumulative_sum = np.cumsum(contributions)

# Plot contributions and cumulative sum
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(n_values, contributions, label='Contribution of each term')
plt.xlabel('n')
plt.ylabel('Contribution to u(1,1)')
plt.title('Individual Contributions')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(n_values, cumulative_sum, marker='o', linestyle='-', label='Cumulative sum')
plt.xlabel('n')
plt.ylabel('Sum of terms')
plt.title('Convergence of Series')
plt.legend()

plt.tight_layout()
plt.show()
```
